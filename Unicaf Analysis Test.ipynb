```python
# Ensuring that necessary columns are numeric
def convert_to_numeric(df, columns):
    for column in columns:
        df[column] = pd.to_numeric(df[column], errors='coerce')
    return df
```

```python
# Define columns for each table
columns_table_1 = [
    'Applications generated',
    'Applications interested',
    'Customers converted',
    'Average value of Customers',
    'Marketing Spend']

columns_table_2 = [
    'Applications received',
    'Applications converted to customers']

# Convert necessary columns to numeric
table1 = convert_to_numeric(table1, columns_table_1)
table2 = convert_to_numeric(table2, columns_table_2)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Plot distributions of numeric columns in Table 1
def plot_distributions(df, columns, table_name):
    for column in columns:
        plt.figure(figsize=(10, 6))
        sns.histplot(df[column], kde=True)
        plt.title(f'Distribution of {column} in {table_name}')
        plt.show()

plot_distributions(table1, columns_table_1, 'Table 1')
```

```python
# Scatter plots for Table 1
def plot_scatter(df, x_column, y_column, table_name):
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=df, x=x_column, y=y_column)
    plt.title(f'Scatter plot of {x_column} vs {y_column} in {table_name}')
    plt.show()

plot_scatter(table1, 'Applications generated', 'Customers converted', 'Table 1')
plot_scatter(table1, 'Marketing Spend', 'Customers converted', 'Table 1')

# Scatter plots for Table 2
plot_scatter(table2, 'Applications received', 'Applications converted to customers', 'Table 2')
```

```python
# Define a function to calculate conversion rates and cost efficiency
def analyze_campaign_data(df):
    df['Conversion Rate (%)'] = (df['Customers converted'] / df['Applications generated']) * 100
    df['Cost per Customer'] = df['Marketing Spend'] / df['Customers converted']
    return df

# Analyze the campaign data from Table 1
campaign_analysis = analyze_campaign_data(table1.copy())

# Define a function to calculate adviser conversion rates
def analyze_adviser_data(df):
    df['Conversion Rate (%)'] = (df['Applications converted to customers'] / df['Applications received']) * 100
    return df

# Analyze the adviser data from Table 2
adviser_analysis = analyze_adviser_data(table2.copy())

# Summarize the key metrics
campaign_summary = campaign_analysis.describe()
adviser_summary = adviser_analysis.describe()

# Create a concise code to perform the analysis
import matplotlib.pyplot as plt

def plot_campaign_performance(df):
    fig, ax1 = plt.subplots(figsize=(12, 8))

    color = 'tab:blue'
    ax1.set_xlabel('Campaign')
    ax1.set_ylabel('Applications Generated', color=color)
    ax1.bar(df['Campaign'], df['Applications generated'], color=color, alpha=0.6, label='Applications Generated')
    ax1.tick_params(axis='y', labelcolor=color)

    ax2 = ax1.twinx()
    color = 'tab:red'
    ax2.set_ylabel('Conversion Rate (%)', color=color)
    ax2.plot(df['Campaign'], df['Conversion Rate (%)'], color=color, marker='o', label='Conversion Rate (%)')
    ax2.tick_params(axis='y', labelcolor=color)

    fig.tight_layout()
    plt.title('Campaign Performance')
    plt.show()

def plot_adviser_performance(df):
    fig, ax = plt.subplots(figsize=(12, 8))
    advisers = df.sort_values('Conversion Rate (%)', ascending=False)['Adviser']
    conversion_rates = df.sort_values('Conversion Rate (%)', ascending=False)['Conversion Rate (%)']
    ax.barh(advisers, conversion_rates, color='tab:green', alpha=0.6)
    ax.set_xlabel('Conversion Rate (%)')
    plt.title('Adviser Performance')
    plt.show()

# Plot campaign performance
plot_campaign_performance(campaign_analysis)

# Plot adviser performance
plot_adviser_performance(adviser_analysis)

# Show the analyzed data and summaries
campaign_analysis, adviser_analysis, campaign_summary, adviser_summary
```
